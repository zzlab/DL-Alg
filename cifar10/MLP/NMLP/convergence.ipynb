```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
model = 'CMLP'
history = []
for depth in range(1, 13):
    shape = (1024,) * depth + (10,)
    path = ('%s-%s10' % (model, '{}-' * (len(shape) - 1))).format(*shape[:-1])
    test_accuracy, progress = pickle.load(open('models/%s-history' % path, 'rb'))
    history.append(test_accuracy)
    pl.figure()
    pl.title('%d-hidden-layer network loss' % depth)
    pl.xlabel('epoch')
    pl.ylabel('loss')
    training_loss = progress['training_loss']
    validation_loss = progress['validation_loss']
    pl.plot(range(len(training_loss)), training_loss, label='training loss')
    pl.plot(range(len(validation_loss)), validation_loss, label='validation loss')
    pl.legend(framealpha=0)
    pl.figure()
    pl.title('%d-hidden-layer network accuracy' % depth)
    pl.xlabel('epoch')
    pl.ylabel('accuracy')
    training_accuracy = progress['training_accuracy']
    validation_accuracy = progress['validation_accuracy']
    pl.plot(range(len(training_accuracy)), training_accuracy, label='training accuracy')
    pl.plot(range(len(validation_accuracy)), validation_accuracy, label='validation accuracy')
    pl.legend(framealpha=0)
pl.figure()
pl.title('test accuracy')
pl.plot(range(len(history)), history)
pl.xlabel('depth')
pl.ylabel('accuracy')
```
