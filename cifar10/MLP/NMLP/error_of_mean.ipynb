```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
model = 'CMLP'
for depth in range(1, 13):
    shape = (1024,) * depth + (10,)
    path = ('%s-%s10' % (model, '{}-' * (len(shape) - 1))).format(*shape[:-1])
    record = pickle.load(open('models/%s-error-of-mean' % path, 'rb'))
    pl.figure()
    pl.title('%d-hidden-layer network error of mean' % depth)
    pl.xlabel('layer index')
    pl.ylabel('error')
    for phase in ('training', 'validation', 'test'):
        minimum = [error.min() for error in record[phase]]
        maximum = [error.max() for error in record[phase]]
        mean = [error.mean() for error in record[phase]]
        pl.plot(range(depth), minimum, label='minimum error in %s' % phase)
        pl.plot(range(depth), maximum, label='maximum error in %s' % phase)
        pl.plot(range(depth), mean, label='mean of error in %s' % phase)
    pl.legend(framealpha=0)
```
