```python
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
import cPickle as pickle
```

```python
N, D = 1024, 2
mean, std = 0, 1
mean_vector = np.full(D, mean)
covariance_matrix = np.eye(D) * std
data = np.random.multivariate_normal(mean_vector, covariance_matrix, N)
X, Y = data.T
```

```python
pl.plot(X, Y, 'bo')
pl.gca().set_aspect('equal')
```

```python
labels = []
for point in data:
    x, y = point
    if x > 0 and y > 0:
        labels.append(0)
    elif x < 0 and y > 0:
        labels.append(1)
    elif x < 0 and y < 0:
        labels.append(2)
    elif x > 0 and y < 0:
        labels.append(3)
    else:
        print x, y
labels = np.array(labels)

total = 0
for i in range(4):
    n = np.sum(labels == i)
    total += n
    print i, n
```

```python
pickle.dump((data, labels), open('sanity-data', 'wb'))
```
