```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
```

```python
X = np.linspace(-1, 1, 1000)
Y = X + np.random.normal(0, 0.1, X.shape)
pl.plot(X, Y, 'bo')
pl.gca().set_aspect('equal')
data = np.stack((X, Y)).T
pickle.dump(data, open('sanity-data', 'wb'))
```
