```python
import numpy as np
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
prefix = 'tuned-ReLU-resnet'
path = 'models/%s-post-activations' % prefix
activations = pickle.load(open(path, 'rb'))
```

```python
INDEX = 0
for phase in 'test', :
    for index, activation in enumerate(activations[phase]):
        pl.figure()
        print len(np.swapaxes(activation, 0, 1)[INDEX].flatten())
        pl.hist(
            np.swapaxes(activation, 0, 1)[INDEX].flatten(),
            bins   = 256,
            normed = True,
            label  = '%s convolution layer %d filter %d' % (phase, index, INDEX)
        )
        pl.legend(framealpha=0)
```
