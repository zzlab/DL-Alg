```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
_, progress = pickle.load(open('history/NIN-DReLU-Adam'))
```

```python
pl.title('loss')
pl.plot(range(len(progress['training_loss'])), progress['training_loss'])
pl.plot(range(len(progress['validation_loss'])), progress['validation_loss'])
```

```python
pl.title('accuracy')
pl.plot(range(len(progress['training_accuracy'])), progress['training_accuracy'])
pl.plot(range(len(progress['validation_accuracy'])), progress['validation_accuracy'])
```
