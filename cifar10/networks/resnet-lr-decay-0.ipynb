```python
import cPickle as pickle
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
activation = 'ReLU'
accuracies = []
progresses = []
for epoch in range(10, 50, 2):
    file_path = 'resnet_lr_decay_epoch_0_grid_search/%s-decay-at-epoch-%d' % (activation, epoch)
    accuracy, progress = pickle.load(open(file_path, 'rb'))
    accuracies.append(accuracy)
    progresses.append(progress)
```

```python
pl.plot(range(len(accuracies)), accuracies)
```

```python
progresses[0].keys()
```

```python
keys = (
    'validation_accuracy',
    'training_loss',
    'validation_loss',
    'training_accuracy'
)
```

```python
for key in keys:
    for progress in progresses:
        pl.figure()
        history = progress[key]
        pl.plot(range(len(history)), history)
        pl.title(key)
```
