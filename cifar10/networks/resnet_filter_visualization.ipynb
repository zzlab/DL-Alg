```python
import numpy as np
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
import sys
sys.path.append('../../nn')
```

```python
_, args, _ = pickle.load(open('models/tuned-resnet-statistics-parameter', 'rb'))
```

```python
filters = args['convolution0_weight'].asnumpy()
for i in range(len(filters)):
    pl.figure()
    pl.axis('off')
    pl.grid(True)
    pl.imshow(filters[i], interpolation='none')
    pl.legend()
```

```python
channel_filters = filters.reshape((48, 3, 3))
for i in range(len(filters)):
    pl.figure()
    pl.axis('off')
    pl.title('filter %d channel %d' % (i // 3, i % 3))
    pl.imshow(channel_filters[i], cmap=pl.gray(), interpolation='none')
```
