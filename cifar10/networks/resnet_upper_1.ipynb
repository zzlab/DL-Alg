```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

lower = 0; upper = 2;

```python
for activation in ('BNReLU', 'BNDReLU'):
    for kernel in ('single', 'double'):
        path = 'history/residual-%d-%s-%s-Adam-history' % (3, activation, kernel)
        _, progress = pickle.load(open(path, 'rb'))
        pl.plot(
            range(len(progress['validation_accuracy'])),
            progress['validation_accuracy'],
            label='%d-%s-%s' % (3, activation, kernel)
        )
for activation in ('BNReLU', 'BNDReLU'):
    path = 'history/residual-6-%s-single-Adam-history' % activation
    _, progress = pickle.load(open(path, 'rb'))
    pl.plot(
        range(len(progress['validation_accuracy'])),
        progress['validation_accuracy'],
        label='%d-%s-%s' % (6, activation, 'single')
    )
pl.legend(framealpha=0, loc='lower right')
```
