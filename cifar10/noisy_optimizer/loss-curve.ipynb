```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
SHAPE = (1024,) * 3 + (10,)
ACTIVATION = 'ReLU'
path = 'mlp-%s-%s' % (ACTIVATION, '-'.join(str(shape) for shape in SHAPE))
info_path = '%s-info' % path
loss, training_loss, validation_accuracy = pickle.load(open(info_path, 'rb'))
```

```python
pl.plot(range(len(training_loss)), training_loss)
pl.ylim(0, 4)
```

```python
pl.plot(range(len(validation_accuracy)), validation_accuracy)
pl.ylim(0, 1)
```
