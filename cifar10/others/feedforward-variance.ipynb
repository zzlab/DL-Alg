```python
import matplotlib.pylab as pl
%matplotlib inline
import minpy.numpy as np
from scipy.stats import norm
```

```python
N = 1000
D = [3072, 1024, 512, 256, 128, 64, 16]
X = np.random.normal(0, 1, (N, D[0]))
weights = [np.random.normal(0, (2 / float(D[i])) ** 0.5, (D[i], D[i + 1])) for i in range(len(D) - 1)]
```

```python
BINS = 256
line_space = np.linspace(-4, 4, 1024).asnumpy()
inputs = X
pl.plot(line_space, norm.pdf(line_space, 0, 1), 'ro')
pl.hist(inputs.asnumpy().flatten(), bins=BINS, normed=True)
pl.plot(line_space, norm.pdf(line_space, 0, 1), 'ro')
pl.title('data distribution')
pl.legend()
for index, W in enumerate(weights):
    inputs = np.dot(inputs, W)
    pl.figure()
    pl.title('layer %d distribution' % index)
    pl.plot(line_space, norm.pdf(line_space, 0, 1), 'ro')
    pl.hist(inputs.asnumpy().flatten(), bins=BINS, normed=True)
    pl.legend()
    inputs = np.maximum(inputs, 0)
```
