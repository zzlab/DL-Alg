```python
import numpy as np
import matplotlib.pylab as pl
%matplotlib inline
```

```python
std_space = np.linspace(1E-10, 3)
```

```python
E_Y_table = []
for std in std_space:
    X = np.random.normal(0, std, 10000)
    Y = np.maximum(0, X)
    E_Y = np.mean(Y)
    E_Y_table.append(E_Y)
pl.plot(std_space, E_Y_table)
pl.xlabel('std')
pl.ylabel('E(X)')
pl.gca().set_aspect('equal', adjustable='box')
```

```python
E_2_Y_table = []
for std in std_space:
    X = np.random.normal(0, std, 10000)
    Y = np.maximum(0, X)
    E_2_Y = np.mean(Y) ** 2
    E_2_Y_table.append(E_2_Y)
pl.plot(std_space, E_2_Y_table)
pl.plot(std_space, 0.16 * std_space ** 2)
pl.xlabel('std')
pl.ylabel('E2(X)')
pl.xlim(0, np.max(std_space))
pl.gca().set_aspect('equal', adjustable='box')
```
