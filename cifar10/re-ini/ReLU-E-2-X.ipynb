```python
import numpy as np
import matplotlib.pylab as pl
%matplotlib inline
```

```python
std_space = np.linspace(1E-10, 4)
```

```python
E_Y_table = []
for std in std_space:
    X = np.random.normal(0, std, 10000)
    Y = np.maximum(0, X)
    E_Y = np.mean(Y)
    E_Y_table.append(E_Y)
pl.plot(std_space, E_Y_table)
pl.xlabel('std')
pl.ylabel('E(X)')
pl.gca().set_aspect('equal')
```

```python
E_Y_table = []
X = np.random.uniform(-1, 1, 10000)
std_X = np.std(X)
for std in std_space:
    rescaled = X * std / std_X 
    Y = np.maximum(0, rescaled)
    E_Y = np.mean(Y)
    E_Y_table.append(E_Y)
pl.plot(std_space, E_Y_table)
pl.xlabel('std')
pl.ylabel('E(X)')
pl.gca().set_aspect('equal')
```

```python
E_Y_table = []
X = np.random.standard_cauchy(10000)
std_X = np.std(X)
for std in std_space:
    rescaled = X * std / std_X 
    Y = np.maximum(0, rescaled)
    E_Y = np.mean(Y)
    E_Y_table.append(E_Y)
pl.plot(std_space, E_Y_table)
pl.xlabel('std')
pl.ylabel('E(X)')
pl.gca().set_aspect('equal')
```

```python
# validate the proportionality of V(f(Y)) to V(Y)
std_Y_table = []
for std in std_space:
    X = np.random.normal(0, std, 10000)
    Y = np.maximum(0, X)
    std_Y = np.std(Y)
    std_Y_table.append(std_Y)
pl.plot(std_space, std_Y_table, label='empirical')
ratio = 0.4
pl.plot(std_space, (0.5 - ratio ** 2) ** 0.5 * std_space, label='theoretical')
pl.xlabel('std')
pl.ylabel('E(X)')
pl.legend(framealpha=0, loc='lower right')
pl.gca().set_aspect('equal')
```

```python
# experiment on distributions that are not 0-centered
E_Y_table = []
for std in std_space:
    X = np.random.normal(3, std, 10000)
    Y = np.maximum(0, X)
    E_Y = np.mean(Y)
    E_Y_table.append(E_Y)
pl.plot(std_space, E_Y_table)
pl.xlabel('std')
pl.ylabel('E(Y)')
pl.gca().set_aspect('equal')
```
