```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
history = pickle.load(open('dynamic-rescaling-loss-history', 'rb'))
```

```python
pl.plot(range(len(history)), history)
pl.ylim(0, 3)
```

```python
length = 100
for i in range(9):
    iteration = (i + 1) * 1000 - 1
    start = iteration - length
    end = iteration + length
    period = history[start : end]
    pl.figure()
    pl.title('rescaling #%d' % i)
    pl.plot(range(start, end), period)
    pl.xlim(start, end)
    pl.ylim(min(period), max(period))
```
