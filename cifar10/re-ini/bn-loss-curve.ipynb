```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
ACTIVATION = ('ReLU',)
SHAPES = ('1024',) * 4 + ('10',)
configuration = '-'.join(ACTIVATION + SHAPES)
_, history = pickle.load(open('model/MLP-%s-history' % configuration, 'rb'))
bn_configuration = 'BN' + '-'.join(ACTIVATION + SHAPES)
_, bn_history = pickle.load(open('model/MLP-%s-history' % bn_configuration, 'rb'))
```

```python
for key, value in history.items():
    if key != 'epoch':
        pl.figure()
        pl.title(key)
        pl.plot(range(len(value)), value, label='normal')
        pl.plot(range(len(bn_history[key])), bn_history[key], label='bn')
        pl.grid()
        pl.legend(framealpha=0)
```
