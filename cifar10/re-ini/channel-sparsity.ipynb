```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
loss, sparsity = pickle.load(open('channel-sparsity', 'rb'))
```

```python
ITERATIONS = 15000
```

```python
pl.plot(range(ITERATIONS), loss)
pl.ylim(0, 3)
pl.grid()
```

```python
for key, value in sparsity.items():
    pl.plot(range(ITERATIONS), value, label=key)
pl.legend(framealpha=0)
pl.grid()
```
