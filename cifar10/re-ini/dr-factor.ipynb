```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
import numpy as np
```

```python
history = pickle.load(open('dr-factor-history', 'rb'))
```

```python
ITERATIONS = 10
for i, h in enumerate(zip(*history)):
    pl.plot(range(ITERATIONS), h, label='affine layer %d' % i)
pl.grid()
pl.legend(framealpha=0)
```

```python
backs = [np.prod(np.array(h)) for h in history]
```

```python
pl.plot(range(ITERATIONS), backs)
```
