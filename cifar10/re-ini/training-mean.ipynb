```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
import numpy as np
```

```python
path = 'centered-weight-checkpoints'
step = 1
total_epochs = 1000
affines = 5
```

```python
for iteration in range(1000, total_epochs + 1, step):
    data = pickle.load(open('%s/iteration-%d' % (path, iteration)))
    for i in range(affines):
        values = data['affine%d' % i]
        mean = values.mean()
        pl.figure()
        pl.hist(values.flatten(), bins=1024, normed=True)
        pl.title('affine {} shape {} mean {}'.format(i, values.shape, mean))
```

```python
for iteration in range(1, total_epochs + 1, step):
    data = pickle.load(open('%s/iteration-%d' % (path, iteration)))
    means = []
    for i in range(affines):
        values = data['affine%d' % i]
        mean = values.mean()
        means.append(mean)
    pl.figure()
    pl.plot(range(len(means)), means)
    pl.legend()
```

```python
affine_means = {'affine%d' % i : [] for i in range(affines)}
# for iteration in range(1, total_epochs + 1, step):
for iteration in range(1, 100, step):
    data = pickle.load(open('%s/iteration-%d' % (path, iteration)))
    for i in range(affines):
        label = 'affine%d' % i
        values = data[label]
        mean = values.mean()
        affine_means[label].append(mean)
for key, value in affine_means.items():
    pl.figure()
    pl.plot(range(len(value)), value)
    pl.title(key)
    pl.xlabel('iteration')
    pl.ylabel('mean')
```
