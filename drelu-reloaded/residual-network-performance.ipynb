```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
for N in (3, 5, 7, 9, 18):
    pl.figure()
    accuracy, progress = pickle.load(open('info/residual-network-n-%d-activate-relu-times-1' % N, 'rb'))
    pl.plot(range(len(progress['validation_accuracy'])), progress['validation_accuracy'], label='single relu %f' % accuracy)
    accuracy, progress = pickle.load(open(
        '/home/alex/experiment/fast_weight/cv/hyper-network/info/triple-state-normal-residual-network-%d' % N,
        'rb'
    ))
    pl.plot(range(len(progress['validation_accuracy'])), progress['validation_accuracy'], label='double relu %f' % accuracy)
    accuracy, progress = pickle.load(open('info/residual-network-n-%d-activate-drelu-times-1' % N, 'rb'))
    pl.plot(range(len(progress['validation_accuracy'])), progress['validation_accuracy'], label='single drelu %f' % accuracy)
    accuracy, progress = pickle.load(open('info/residual-network-n-%d-activate-drelu-times-2' % N, 'rb'))
    pl.plot(range(len(progress['validation_accuracy'])), progress['validation_accuracy'], label='double drelu %f' % accuracy)
    pl.title('%d layers' % (6 * N + 2))
    pl.xlabel('epoch')
    pl.ylabel('validation accuracy')
    pl.legend(framealpha=0, loc='best')
    pl.ylim(0.2, 1.0)
    pl.grid()
```
